```python
import pandas as pd
```

```python
dataset =pd.read_csv("insurance.csv")
```

```python
dataset
```

```python
dataset = pd.get_dummies(dataset, drop_first = True)
```

```python
dataset
```

```python
dataset.columns
```

```python
independent = dataset[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
```

```python
dependent =dataset[['charges']]
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVR
```

```python
# kernel poly is taking time so i removed from the list
param_grid = {'kernel':['rbf','sigmoid','linear'], 'C':[100,1000,2000,3000],'gamma':['auto','scale']}
grid = GridSearchCV(SVR(), param_grid, cv=5, refit = True, verbose = 3,n_jobs=-1)
grid.fit(independent, dependent)


```

```python
re=grid.cv_results_
print("TheR_scorevalueforbestparameter{}:".format(grid.best_params_))
```

```python
table=pd.DataFrame.from_dict(re)
```

```python
table
```

```python
age_input=float(input("Age:"))
bmi_input=float(input("BMI:"))
children_input=float(input("Children:"))
sex_male_input=int(input("Sex Male 0 or 1:"))
smoker_yes_input=int(input("Smoker Yes 0 or 1:"))

Future_Prediction=grid.predict([[age_input,bmi_input,children_input,sex_male_input,smoker_yes_input]])
#␣ →change the paramter,play with it. print("Future_Prediction={}".format(Future_Prediction))
```
